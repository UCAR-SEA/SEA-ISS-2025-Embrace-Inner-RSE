# Helping Scientists Embrace their Inner Research Software Engineer (RSE) and Working Together with the Community Earth System Model (CESM) RSEs to improve CESM Science

## From a Land Component Model (CTSM) Perspective (CTSM is the (Community Terrestrial System Model)

**Author:** [Erik Kluzek](mailto:erik@ucar.edu)
**Keywords:** Research-Software-Engineering, Software-Development-Methods, CESM, CTSM, Earth-System-Model, Land-Model

# Abstract:

The Community Earth System Model {cite}`githubCESM,webCESM` is science expressed in Software, which means everyone who touches CESM code to work on the science -- is doing RSE work. There is a list of challenges that CESM RSEs have with both bringing in new science for CESM and ensuring CESM is well tested, robust, flexible as well as having the correct science. As such CESM scientists working in CESM code need to work alongside CESM RSEs to adopt RSE practices while their science is being developed. For CESM to be useful for scientists it needs to be: well-tested, robust, reproducible, flexible and have assurance that the science is implemented correctly. This is needed both for the science and the software of CESM. Helping  scientists embrace their inner RSE is an effort to keep CESM as a leading model in Earth System Science research, by having more people who work with the CESM code adopt good coding practices. In this talk I will specifically talk about RSE practices for the Land Model component of CESM, the Community Terrestrial Science Model {cite}`githubCTSM`. I will present on the RSE challenges we have in CTSM and how to solve them we need to involve and educate CTSM scientists to improve our RSE practices. I will also present regarding the CTSM RSE team’s efforts to educate the scientists and on some improvements to our RSE practices that we’ve been implementing on our CTSM Software Development.


# Introduction

CESM has been a multidecadal project for studying the Earth's Climate System. It's a large software system of millions of lines of code with interdisciplinary experts spanning many scientific disciplines from Atmosphere, Hydrosphere, Cryosphere, including RSEs tasked with keeping the software itself working in a sustainable way.
## Problem Statement

In CTSM as a Community model, code contributions come from various scientists from all over the nation as well as all over the globe. These contributors to the code come from various scientific backgrounds. They are often graduate students in a scientific discipline, or can be professors, or scientific researchers. In most cases they are not trained in software engineering and often have little formal training in programming. We have a small team of RSE Professionals tasked with managing the software. However, the small but mighty CTSM SE team does not have the capacity to rewrite or refactor most of incoming contributions. Nor is there capacity to refactor the entire working  set of code to modernize or remove technical debt.
- Scientists contribute a substantial amount of code – untrained in RSE practices
- Scientists need to concentrate on their science
- Encourage better Software Development practices from everyone
- Increasing collaboration between RSEs and code contributors
- Part of this talk is from a presentation to our scientists
- This is our story of adding lines of collaboration
- Bad code/ Bad RSE practices result in science problems. 

In the development of CTSM scientists contribute code as well as Software Engineers. To some degree the two have worked in isolation, with scientists contributing code, and then RSEs working to test, validate and improve it if time is available (see {numref}`Fig-separate` below). A shortcoming here is that RSEs often don't see code updates until late in the development cycle, and can't be involved in improving the code earlier in that development. It also means that scientists contributing code aren't able to utilize modern software development methodologies -- because they simply aren't trained in them.

```{figure} ../assets/Scientist_SE_Contrib.png
:alt: when scientists and RSEs contribute code separately there is less opportunity for collaboration
:name: Fig-separate

Scientists and SE contribute separately

When scientists and RSEs only act independently the result code and coding practices aren't as sustainable.
Photo credit: Carla Martinez (programmer), National Cancer Institute (scientists) from unsplash.com
```

## What might be Solutions to Try?

### How do we Encourage Scientists to improve Coding Practices?

The central question we have then is what can we do to encourage scientists to learn RSE best practices and put them into place. This is a list of things there we are trying.
- Help them to see that poor code impacts the science
- When problems come up examine the causes and how to prevent
- Point out the missing practices that could’ve helped
- Point out the cost of having poor practices and how that’s greater than cost of practices

### How do we Encourage Scientists to improve Coding Practices? Education

Another important point of encouraging scientists to adopt better RSE practices is that they have to be educated on them. Scientists are busy and don't have time to spend inordinate amount of time learning an arbitrary software skill. And we can't expect them to take extensive coursework in Computer Science. Nor can they be expected to search for the latest skills and practices to try out in their coding. The main education they need to pursue is in their scientific discipline and keeping up with the latest advancements in their field. However, if RSEs can work on giving that education, and make it easy to learn, and only focus on the most important practices -- scientists can pick up these skills. These are the kind of educational practices we are starting to work on in order to accomplish this.

- Educate them on the most important practices
- RSE team actively doing and experts in the practices
- Have education resources available (tutorials, documentation etc.)
- Use metaphors and non-technical terms
- Encourage people that try and provide help
- Education on the practices needs to continue to be done
- Introduce one practice at a time
- Make most important things easy

### What Does NOT Work in a Community model?

In my experience in working with Scientists over several decades, there's been a list of things I've tried, or seen attempted to be tried that have not been successful. Most of this list really just has to do with bad ways of treating people and as such ways we can't expect would work for anyone. But, as in the previous section it's important to have realistic expectations of what can be expected of community scientists contributing code to a community model. We can't realistically expect that they will have a Computer Science educational background nor extensive knowledge on the subject. So being able to let go of unrealistic demands is one key to doing this well.
- Demanding – “you must do ____”
- Expecting – “you should know ____”
- Shaming – “(condescendingly) I told you so – you should have done ____”
- Technical – “Read this textbook on Software Development Methodology”
- Claiming – “We HAVE to be exactly like the SE Industry”
- Academic – “Read and digest this study on what it shows”
- Overwhelm – “Do ALL of these 10 things at once”
- No relevance – “Do these RSE things – forget about the science”
- Wishing – “Maybe someone will do this on their own”

### Summary: What Does Work in a Community model?

As such we think that these are the most important things to try to do in this space.
- Demonstrate costs for bad practices
 - Demonstrate practices
 - Educate on practices
 - Work with people with the highest interests in RSE work
